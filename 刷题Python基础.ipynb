```python
#字符串
s = "Hello, World!"
#查改增删用法


#---1.查询------
#字符串切片
substring = s[0:5]  # 取字符串的前5个字符，记住不包含索引5
print(substring)  # 输出: Hello

#字符串索引
char = s[7]  # 取字符串中索引为7的字符
print(char)  # 输出: W


#负数索引
last_char = s[-1]  # 取字符串的最后一个字符
print(last_char)  # 输出: !


#---2.修改------
#字符串是不可变的，所以不能直接修改字符串中的字符，但可以通过重新赋值replace来修改字符串
#例如，将字符串中的 "World" 替换为 "Python"
new_string = s.replace("World", "Python")
print(new_string)  # 输出: Hello, Python!
```

```python
#内置函数

#Max
#(1)直接比较多个数值，返回最大值
max_value = max(1, 5, 3, 9, 2) 
print(max_value)  # 输出: 9

#(2)比较字符串，返回字典序最大的字符串
max_string = max("apple", "banana", "cherry") 
print(max_string)  # 输出: cherry

#(3)使用key参数比较字符串长度，返回最长的字符串
max_len_string = max("apple", "banana", "cherry", key=len) 
print(max_len_string)  # 输出: banana

```

```python
#链表
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

#----1.创建链表-------
#创建一个链表 1 -> 2 -> 3
node1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)

node1.next = node2
node2.next = node3

#将node1作为链表的头节点
head = node1

#----2.查询链表-------
#遍历链表并打印每个节点的值
current = head
while current:
    print(current.val)  # 输出: 1, 2, 3
    current = current.next
```

```python
# 在算法中，这个 “堆 (Heap)”（也叫优先队列）和内存管理的堆区不是一个概念
# 堆是一种特殊的树形数据结构
# 满足堆的性质：在最大堆中，父节点的值总是大于等于子节点的值；在最小堆中，父节点的值总是小于等于子节点的值。
# 在Python中，可以使用内置的heapq模块来实现堆数据结构，提供了最小堆的功能。
# 这个名字其实是 Heap（堆） 和 Queue（队列） 的缩写组合
import heapq

# 创建一个空的最小堆
min_heap = []

# 向堆中插入元素
# 语法：heapq.heappush(heap, item)，其中heap是要插入元素的堆，item是要插入的元素。元素可以是任何可比较的类型，如数字、字符串等
# 例如，向最小堆中插入一些数字：
heapq.heappush(min_heap, 5)
heapq.heappush(min_heap, 2)
heapq.heappush(min_heap, 8)

# 弹出堆顶元素
top = heapq.heappop(min_heap)
print(top)  # 输出: 2 (最小的元素被弹出)


#示例二，插入元组 (优先级, 任务)
task_heap = []  
heapq.heappush(task_heap, (2, "Task A"))
heapq.heappush(task_heap, (1, "Task B"))
heapq.heappush(task_heap, (3, "Task C"))

# 弹出优先级最高的任务-第一优先级
priority, task = heapq.heappop(task_heap)
print(f"Processing {task} with priority {priority}")  # 输出: Processing Task B with priority 1

#打印堆
print(min_heap)  # 输出: [5, 8] (剩余的元素)
print(task_heap)  # 输出: [(2, 'Task A'), (3, 'Task C')] (剩余的任务)


#其他操作
#弹出
#heapq.heappop(heap) 从堆中弹出并返回最小的元素，同时保持堆的性质。
#heapq.heappushpop(heap, item) 将一个元素插入堆中，然后弹出并返回堆中的最小元素。这是一个原子操作，效率比先插入再弹出更高。
#heapq.heapify(x) 将列表x转换成一个堆，原地进行操作
```
