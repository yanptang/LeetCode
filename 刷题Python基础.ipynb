```python
# 数组
arr=[1,2,3,4,5]
matrix=[[1,2,3],[4,5,6],[7,8,9]]

#----1.查询-------
#取数组最后一个元素的用法
last_element = arr[-1]
print(last_element)  # 输出: 5

# 取矩阵最后一行最后一个元素的用法
last_element_matrix = matrix[-1]
print(last_element_matrix)  # 输出: [7, 8, 9]



#----2.切片-------
```

```python
#字符串
s = "Hello, World!"
#查改增删用法


#---1.查询------
#字符串切片
substring = s[0:5]  # 取字符串的前5个字符，记住不包含索引5
print(substring)  # 输出: Hello

#字符串索引
char = s[7]  # 取字符串中索引为7的字符
print(char)  # 输出: W


#负数索引
last_char = s[-1]  # 取字符串的最后一个字符
print(last_char)  # 输出: !


#---2.修改------
#字符串是不可变的，所以不能直接修改字符串中的字符，但可以通过重新赋值replace来修改字符串
#例如，将字符串中的 "World" 替换为 "Python"
new_string = s.replace("World", "Python")
print(new_string)  # 输出: Hello, Python!
```

```python
#内置函数

#Max
#(1)直接比较多个数值，返回最大值
max_value = max(1, 5, 3, 9, 2) 
print(max_value)  # 输出: 9

#(2)比较字符串，返回字典序最大的字符串
max_string = max("apple", "banana", "cherry") 
print(max_string)  # 输出: cherry

#(3)使用key参数比较字符串长度，返回最长的字符串
max_len_string = max("apple", "banana", "cherry", key=len) 
print(max_len_string)  # 输出: banana

```

```python
#特殊用法
#---1.列表推导式------

#快速生成一个包含5个元素的列表，每个元素都是1
res=[1]*5
print(res)  # 输出: [1, 1, 1, 1, 1]

#二维列表推导式，生成一个3x3的矩阵，每个元素都是0
matrix = [[0]*3 for _ in range(3)]
print(matrix)  # 输出: [[0, 0, 0], [0, 0, 0], [0, 0, 0]]

#生成一个包含0到9的平方的列表
squares = [x**2 for x in range(10)]
print(squares)  # 输出: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
```
